##Task V

```python
car_auction_df.boxplot(column='VehOdo')
```

The graph tells us the maximum, 75th percentile, 50th percentile, 25th percentile, and minimum values in the associated column, along with outlier values.

```python
car_auction_df['Size'].value_counts().plot(kind='bar')
```

The barplot shows the Size values on vertical axis and Size values of cars and the associated counts.

```python
sns.histplot(data=car_auction_df['WarrantyCost'],bins=30)
```

The plot shows, in twenty bins, a histogram of counts of the Warranty Cost values that fall within the boundaries of the associated bins.

```python
car_auction_df = car_auction_df.drop(['over_80k_miles'],axis=1)
```

```python
sns.pairplot(data=car_auction_df)
```

```python
sns.pairplot(data=car_auction_df,hue='IsBadBuy')
```

The pair plot shows approximated continuous count values along the identically paired diagonal, and scatter plots of the values on the off-diagonal, with the horizontal axes indicating the column variable and the vertical axes indicating the row variable. 

Particularly bad buys include vehicles that are over 80,000 miles on their odometer, but have a high auction price or high vehicle cost. Other bad buys may include high warranty cost and high vehicle age, low warranty cost with high auction price or high vehicle cost, or high vehicle cost with low warranty cost or low vehicle odometer.
